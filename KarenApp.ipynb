```python
import csv
import numpy as np
import pandas as pd
from sklearn.neural_network import MLPClassifier
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import LabelEncoder
```

```python
data = []
ifile = open('DatosKarenApp.csv', 'rb')
reader = csv.reader(ifile)

rownum = 0
column = 0
for row in reader:
    if rownum == 0:
        header = row
    else:
        colnum = 0
        for col in row:
            data.append(row)
            print '%-8s: %s' % (header[colnum], col)
            colnum += 1
            column += 1
    rownum += 1
ifile.close()
print 
```

```python
column
```

```python
dataframe = pd.DataFrame(data = data, columns = header)
dataframe
```

```python
trainCount = int(column*0.7)

trainData = dataframe.iloc[:trainCount,:]
testData = dataframe.iloc[trainCount:, :]

xTrainData = trainData.iloc[:,2:]
xTestData = testData.iloc[:,2:]
yTrainData = trainData.iloc[:,2]
yTestData = testData.iloc[:,2]
```

```python
encoder = LabelEncoder()
scaler = StandardScaler()
```

```python
ndxTrainData = xTrainData.values
ndyTrainData = yTrainData.values
ndxTestData = xTestData.values
ndyTestData = yTestData.values
```

```python
encoder.fit(ndxTrainData[:, 4])
ndxTrainData[:, 4] = encoder.transform(ndxTrainData[:, 4])

encoder.fit(ndxTrainData[:, 5])
ndxTrainData[:, 5] = encoder.transform(ndxTrainData[:, 5])

encoder.fit(ndxTrainData[:, 8])
ndxTrainData[:, 8] = encoder.transform(ndxTrainData[:, 8])
ndxTrainData
```

```python
encoder.fit(ndxTestData[:, 4])
ndxTestData[:, 4] = encoder.transform(ndxTestData[:, 4])

encoder.fit(ndxTestData[:, 5])
ndxTestData[:, 5] = encoder.transform(ndxTestData[:, 5])

encoder.fit(ndxTestData[:, 8])
ndxTestData[:, 8] = encoder.transform(ndxTestData[:, 8])
ndxTestData
```

```python
scaler.fit(ndxTrainData)
ndxTrainData = scaler.transform(ndxTrainData)
ndxTrainData
```

```python
scaler.fit(ndxTestData)
ndxTestData = scaler.transform(ndxTestData)
ndxTestData
```

```python
result = 0
for i in range(1, 2):
    clf = MLPClassifier(solver='lbfgs', alpha=1e-5, hidden_layer_sizes=(i,), random_state=1)
    clf.fit(ndxTrainData, ndyTrainData)
    result = clf.score(ndxTestData, ndyTestData)
    
result
```
